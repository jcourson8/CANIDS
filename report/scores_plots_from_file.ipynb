```python
import matplotlib.pyplot as plt

for file_data in file_and_scores_and_lables:
    file_name, scores_labels = file_data
    scores = [score for score, label in scores_labels]
    labels = [label for score, label in scores_labels]

    plt.figure(figsize=(10, 6))
    plt.scatter([i for i, label in enumerate(labels) if label == 0], 
                [score for score, label in scores_labels if label == 0], 
                color='blue', label='Label 0')

    # plot red points (label 1) on top
    plt.scatter([i for i, label in enumerate(labels) if label == 1], 
                [score for score, label in scores_labels if label == 1], 
                color='red', label='Label 1')
    plt.xlabel('Sample Index')
    plt.ylabel('Score')
    plt.title(f'Scores for {file_name}')
    plt.show()
```
